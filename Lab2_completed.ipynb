```python
# Practice the use of matplotlib
import numpy as np
import matplotlib.pyplot as plt
# Create a data set 
x = np.linspace(2, 18, 100)
print(x)
# Create a figure using plt.figure
fig = plt.figure()
# Plot the sine wave of x
_ = plt.plot(x, np.sin(x), '-')
# Plot the cosine wave of x
_ = plt.plot(x, np.cos(x), '-')
plt.show()
```

```python
# Create two arrays of 5 arbitrary numbers 
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1, 3, 6, 9, 12])
y = np.array([2, 3, 4, 5, 6])
x1 = np.array([0, 2, 4, 8, 10])
y1 = np.array([1, 2, 3, 4, 5])
# Plot two bar charts using plt.bar and the arrays you've created above
# Label them "Example one" and "example two"
fig = plt.figure(figsize = (20, 6))
plt.bar(x1, y1, label="Example one", color='b')
plt.bar(x, y, label="Example two", color='g')
plt.legend()
# Label x axis "Bar number"
plt.xlabel("Bar number")
# Label y axis "Bar height"
plt.ylabel("Bar height")
# Set title "Bar Graph"
plt.title("Bar Graph")
# Show the chart
plt.show()
```

```python
# Histogram Code 
import numpy as np
import matplotlib.pyplot as plt
# Create an array of a random normal distribution
x = np.random.laplace(loc=15, scale=1.5, size=1000)
# Create a histrogram using plt hist()
n, bins, patches = plt.hist(x, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
# Set grid for x and y
# set transparency of the grid using alpha=0.5
plt.grid(axis='both', alpha=0.5)
#plt.grid(axis='y', alpha=0.5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Example of Random that looks normal')
plt.text(23, 45, r'$\mu=15, b=3$')
maxfreq = n.max()
# Set a clean upper y-axis limit.
plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
         
```

```python
# Practice the use of matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Create a data set. x is actually time.
x = np.linspace(2, 15, 100)
# Print the data
print(x)
# Create a figure using plt.figure
fig = plt.figure(figsize=(7,9))
# Plot the sine wave of x
amplitude = np.sin(x)
plt.plot(x, amplitude, '-')
# Plot the cosine wave of x
amplitude = np.cos(x)
plt.plot(x, amplitude, '-')
plt.show()
```
